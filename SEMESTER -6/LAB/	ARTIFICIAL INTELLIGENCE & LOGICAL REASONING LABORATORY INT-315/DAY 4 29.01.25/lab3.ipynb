```python
#BFS WITH GOAL NODE

from collections import deque

def get_graph_from_user():
    graph = {}
    n = int(input("Enter the number of Nodes: "))

    for _ in range(n):
        node = input("Enter the node name: ")
        neighbors = input(f"Enter the neighbors of {node} (comma-separated): ").split(",")
        neighbors = [neighbor.strip() for neighbor in neighbors]  
        graph[node] = neighbors
    return graph


def BFS(graph, start, goal):
    dq = deque()
    visited = {node: False for node in graph}  

    visited[start] = True
    dq.append(start)

    while dq:
        current = dq.popleft()  
        print(current, end=" -> ")
        
        if current == goal:
            print("Goal node reached!")
            return

        for neighbor in graph[current]:  
            if not visited[neighbor]:
                visited[neighbor] = True
                dq.append(neighbor)

graph = get_graph_from_user()
print("\nGraph:", graph)

start_node = input("Enter the starting node: ")  
goal_node = input("Enter the goal node: ")
BFS(graph, start_node, goal_node)
```

```python

#DFS WITH ALL PATH FROM THE STARING NODE AND WITH GOAL NODE

from collections import deque

def get_graph_from_user():
    graph = {}
    n = int(input("Enter the number of Nodes: "))

    for _ in range(n):
        node = input("Enter the node name: ")
        neighbors = input(f"Enter the neighbors of {node} (comma-separated): ").split(",")
        neighbors = [neighbor.strip() for neighbor in neighbors if neighbor.strip()]
        graph[node] = neighbors
    return graph

def BFS_all_paths(graph, start, goal):
    queue = deque([(start, [start])])  # (node, path)
    all_paths = []

    while queue:
        node, path = queue.popleft()

        if node == goal:
            all_paths.append(path)
        else:
            for neighbor in graph.get(node, []):
                if neighbor not in path:  # Avoid cycles
                    queue.append((neighbor, path + [neighbor]))

    return all_paths

graph = get_graph_from_user()
print("\nGraph:", graph)

start_node = input("Enter the starting node: ")
goal_node = input("Enter the goal node: ")

print(f"\nAll possible paths from {start_node} to {goal_node}:")
paths = BFS_all_paths(graph, start_node, goal_node)

if paths:
    for path in paths:
        print(" -> ".join(path))
else:
    print("No path found from start to goal.")
```
