```python
from itertools import permutations

def solve_cryptarithmetic(addends, result):
    # Combine all words and extract unique letters
    all_letters = set(''.join(addends + [result]))
    
    # Check if we have too many unique letters
    if len(all_letters) > 10:
        return "Too many unique letters"
    
    # Get the first letter of each word to handle the no-leading-zero constraint
    first_letters = set(word[0] for word in addends + [result])
    
    solutions = []  # List to store all solutions

    # Generate all possible digit permutations
    for perm in permutations(range(10), len(all_letters)):
        # Create a letter-to-digit mapping
        letter_map = dict(zip(all_letters, perm))
        
        # Check the no-leading-zero constraint
        if any(letter_map[letter] == 0 for letter in first_letters):
            continue
        
        # Convert words to numbers
        numbers = [int(''.join(str(letter_map[c]) for c in word)) for word in addends]
        result_num = int(''.join(str(letter_map[c]) for c in result))
        
        # Check if the equation is valid
        if sum(numbers) == result_num:
            solutions.append(letter_map.copy())  # Add a copy of the solution
    
    return solutions if solutions else "No solution found"

# Example usage
addends = ["TWO", "TWO"]
result = "FOUR"

solutions = solve_cryptarithmetic(addends, result)

if isinstance(solutions, list):
    print(f"Found {len(solutions)} solution(s):")
    for i, solution in enumerate(solutions, 1):
        print(f"\nSolution {i}:")
        for letter, digit in solution.items():
            print(f"{letter} = {digit}")
        # Print the equation with the solution
        print("Equation:")
        for word in addends + [result]:
            num = ''.join(str(solution[c]) for c in word)
            print(f"{word} = {num}")
        if i < len(solutions):
            print("\n---")
else:
    print(solutions)
```
