```python
#WATER JUG PROBLEM

from collections import deque

# Function to find the minimum operations to obtain
# d liters in one jug
def min_steps(m, n, d):
    if d > max(m, n):
        return -1

    # Queue for BFS: (jug1, jug2, steps)
    q = deque([(0, 0, 0)])

    # For tracking the visited states
    visited = [[False] * (n + 1) for _ in range(m + 1)]
    visited[0][0] = True

    while q:
        jug1, jug2, steps = q.popleft()

        if jug1 == d or jug2 == d:
            return steps

        # 1: Fill jug1
        if not visited[m][jug2]:
            visited[m][jug2] = True
            q.append((m, jug2, steps + 1))

        # 2: Fill jug2
        if not visited[jug1][n]:
            visited[jug1][n] = True
            q.append((jug1, n, steps + 1))

        # 3: Empty jug1
        if not visited[0][jug2]:
            visited[0][jug2] = True
            q.append((0, jug2, steps + 1))

        # 4: Empty jug2
        if not visited[jug1][0]:
            visited[jug1][0] = True
            q.append((jug1, 0, steps + 1))

        # 5: Pour jug1 into jug2
        pour1to2 = min(jug1, n - jug2)
        if not visited[jug1 - pour1to2][jug2 + pour1to2]:
            visited[jug1 - pour1to2][jug2 + pour1to2] = True
            q.append((jug1 - pour1to2, jug2 + pour1to2, steps + 1))

        # 6: Pour jug2 into jug1
        pour2to1 = min(jug2, m - jug1)
        if not visited[jug1 + pour2to1][jug2 - pour2to1]:
            visited[jug1 + pour2to1][jug2 - pour2to1] = True
            q.append((jug1 + pour2to1, jug2 - pour2to1, steps + 1))

    return -1

if __name__ == "__main__":

    # jug1 = 4 litre, jug2 = 3 litre
    m, n, d = 4, 3, 2
    print(f"Minimum Number of steps required : {min_steps(m, n, d)}")
```

```python
from collections import deque

def water_jug_bfs(a, b, target):
    visited = set()
    queue = deque()
    queue.append((0, 0))  # Start with both jugs empty

    while queue:
        x, y = queue.popleft()

        if (x, y) in visited:
            continue
        visited.add((x, y))

        # ✅ Goal check
        if x == target or y == target:
            return True

        # All possible operations
        next_states = [
            (a, y),     # Fill jug A
            (x, b),     # Fill jug B
            (0, y),     # Empty jug A
            (x, 0),     # Empty jug B
            (x - min(x, b - y), y + min(x, b - y)),  # A → B
            (x + min(y, a - x), y - min(y, a - x))   # B → A
        ]

        for state in next_states:
            if state not in visited:
                queue.append(state)

    return False
a = 4  # capacity of jug A
b = 3  # capacity of jug B
target = 2

print(water_jug_bfs(a, b, target))  # Output: True
```

```python
#N QUEEN PROBLEM FOR MULTIPLE SOLUTIONS


def is_safe(board, row, col, N):
    for i in range(row):
        if board[i][col]:
            return False
    for i, j in zip(range(row - 1, -1, -1), range(col - 1, -1, -1)):
        if board[i][j]:
            return False
    for i, j in zip(range(row - 1, -1, -1), range(col + 1, N)):
        if board[i][j]:
            return False
    return True

def solve_nq(board, solutions, row, N, count):
    if row == N:
        soln = []
        for i in range(N):
            for j in range(N):
                if board[i][j]:
                    soln.append(j + 1)
                    break  # Only one queen per row
        solutions.append(soln)
        count[0] += 1
        return

    for col in range(N):
        if is_safe(board, row, col, N):
            board[row][col] = 1
            solve_nq(board, solutions, row + 1, N, count)
            board[row][col] = 0

def main():
    N = int(input("Enter the number of Queens (N): "))
    board = [[0] * N for _ in range(N)]
    solutions = []
    count = [0]  # Use a list to modify the count within the function
    solve_nq(board, solutions, 0, N, count)

    for soln in solutions:
        print("[", end="")
        for col in soln:
            print(f"\t{col}", end="")
        print("]")
    print(f"Total solutions: {count[0]}")


if __name__ == "__main__":
    main()
```

```python
def is_safe(node, color, graph, colors, c):
    for neighbor in graph[node]:
        if colors[neighbor] == c:
            return False
    return True

def solve(node, graph, m, colors):
    if node == len(graph):
        return True

    for c in range(1, m + 1):
        if is_safe(node, c, graph, colors, c):
            colors[node] = c
            if solve(node + 1, graph, m, colors):
                return True
            colors[node] = 0  # backtrack

    return False

def graph_coloring(graph, m):
    colors = [0] * len(graph)
    if solve(0, graph, m, colors):
        return colors
    else:
        return "Not possible with {} colors".format(m)
graph = {
    0: [1, 2],
    1: [0, 2],
    2: [0, 1, 3],
    3: [2]
}
m = 3  # max 3 colors

print(graph_coloring(graph, m))
```

```python
#crypt arithmetic problem with multiple solutions
from itertools import permutations

def solve_cryptarithmetic(addends, result):
    # Combine all words and extract unique letters
    all_letters = set(''.join(addends + [result]))
    
    # Check if we have too many unique letters
    if len(all_letters) > 10:
        return "Too many unique letters"
    
    # Get the first letter of each word to handle the no-leading-zero constraint
    first_letters = set(word[0] for word in addends + [result])
    
    solutions = []  # List to store all solutions

    # Generate all possible digit permutations
    for perm in permutations(range(10), len(all_letters)):
        # Create a letter-to-digit mapping
        letter_map = dict(zip(all_letters, perm))
        
        # Check the no-leading-zero constraint
        if any(letter_map[letter] == 0 for letter in first_letters):
            continue
        
        # Convert words to numbers
        numbers = [int(''.join(str(letter_map[c]) for c in word)) for word in addends]
        result_num = int(''.join(str(letter_map[c]) for c in result))
        
        # Check if the equation is valid
        if sum(numbers) == result_num:
            solutions.append(letter_map.copy())  # Add a copy of the solution
    
    return solutions if solutions else "No solution found"

# Example usage
addends = ["TWO", "TWO"]
result = "FOUR"

solutions = solve_cryptarithmetic(addends, result)

if isinstance(solutions, list):
    print(f"Found {len(solutions)} solution(s):")
    for i, solution in enumerate(solutions, 1):
        print(f"\nSolution {i}:")
        for letter, digit in solution.items():
            print(f"{letter} = {digit}")
        # Print the equation with the solution
        print("Equation:")
        for word in addends + [result]:
            num = ''.join(str(solution[c]) for c in word)
            print(f"{word} = {num}")
        if i < len(solutions):
            print("\n---")
else:
    print(solutions)
```

```python
#min max iterative
import math

def minimax(cur_depth, node_index, max_turn, scores, target_depth, b_factor):
    if cur_depth == target_depth:
        return scores[node_index]
    if max_turn:
        largest = None
        for i in range(b_factor):
            index = node_index * b_factor + i
            if index >= len(scores):
                continue
            cur = minimax(cur_depth + 1, index, False, scores, target_depth, b_factor)
            if largest == None or cur > largest:
                largest = cur
        return largest
    else:
        smallest = None
        for i in range(b_factor):
            index = node_index * b_factor + i
            if index >= len(scores):
                continue
            cur = minimax(cur_depth + 1, index, True, scores, target_depth, b_factor)
            if smallest == None or cur < smallest:
                smallest = cur
        return smallest

scores = [int(s) for s in input("Enter the scores:").split()]
b_factor = int(input("enter the branching factor:"))
player = int(input("maximizer or minimizer?(Enter 1 for maximizer and 0 for minimizer):"))
print(player)
tree_depth = math.ceil(math.log(len(scores), b_factor))
print("the optimal value is:", end="")
print(minimax(0, 0, player, scores, tree_depth, b_factor))
```

```python
from collections import deque

def orangesRotting(grid):
    rows, cols = len(grid), len(grid[0])
    queue = deque()
    fresh = 0

    # Step 1: Initialize queue with all rotten oranges
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == 2:
                queue.append((r, c, 0))  # (row, col, time)
            elif grid[r][c] == 1:
                fresh += 1

    directions = [(0,1), (1,0), (0,-1), (-1,0)]  # right, down, left, up
    time = 0

    # Step 2: BFS to rot adjacent fresh oranges
    while queue:
        r, c, t = queue.popleft()
        time = max(time, t)
        for dr, dc in directions:
            nr, nc = r + dr, c + dc
            if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == 1:
                grid[nr][nc] = 2  # rot the orange
                fresh -= 1
                queue.append((nr, nc, t + 1))

    # Step 3: Check if any fresh orange is left
    return time if fresh == 0 else -1
```

```python
def water_connection(n, pipes):
    from collections import defaultdict

    start = [0] * (n + 1)
    end = [0] * (n + 1)
    diam = [0] * (n + 1)

    for a, b, d in pipes:
        start[a] = b
        diam[a] = d
        end[b] = 1  # b has incoming pipe

    result = []

    for i in range(1, n + 1):
        # i is a source (no incoming pipe and has outgoing)
        if end[i] == 0 and start[i] != 0:
            cur = i
            min_d = float('inf')
            # Follow the pipe path until the end
            while start[cur] != 0:
                min_d = min(min_d, diam[cur])
                cur = start[cur]
            result.append((i, cur, min_d))

    return result
n = 9
pipes = [
    (7, 4, 98),
    (5, 9, 72),
    (4, 6, 10),
    (2, 8, 22),
    (9, 7, 17)
]

print(water_connection(n, pipes))
```
