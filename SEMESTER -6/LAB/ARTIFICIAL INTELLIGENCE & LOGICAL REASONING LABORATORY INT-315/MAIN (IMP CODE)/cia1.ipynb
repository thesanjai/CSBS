```python
#MAGIC SQUARE

def magic_square(matrix):
    n = len(matrix)
    magic_sum = n*(n**2+1)//2
    row_sums = [0] * n
    col_sums = [0] * n
    diag_sum = 0
    diag2_sum = 0
    
    for i in range(n):
        for j in range(n):
            row_sums[i] += matrix[i][j]
            col_sums[j] += matrix[i][j]
            if i == j:
                diag_sum += matrix[i][j]
            if i+j == n-1:
                diag2_sum += matrix[i][j]

    if diag_sum != magic_sum or diag2_sum != magic_sum:
        return False
    
    for i in range(n):
        if row_sums[i] != magic_sum or col_sums[i] != magic_sum:
            return False
    return True
    

matrix = [[2,7,6],[9,5,1],[4,3,8]]

if magic_square(matrix):
    print("Yes")
else:
    print("No")
```

```python
def min_time_to_hole(mice, holes):
    # Step 1: Sort both lists
    mice.sort()
    holes.sort()

    # Step 2: Calculate the max distance between assigned pairs
    max_time = 0
    for i in range(len(mice)):
        max_time = max(max_time, abs(mice[i] - holes[i]))

    return max_time

mice = [4, -4, 2]
holes = [4, 0, 5]

print("Minimum maximum time:", min_time_to_hole(mice, holes))
```

```python
#All path finder 

def find_all_paths(graph,start,goal,path=None):
    if path is None:
        path = []
    path = path + [start]
    
    if start not in graph:
        return []

    if start == goal:
        return [path]

    paths =[]
    
    for neighbor in graph[start]:
        if neighbor not in path:
            new_paths = find_all_paths(graph,neighbor,goal,path)
            for p in new_paths:
                paths.append(p)
    return paths

start = input("Enter the Start Node: ")
goal = input("Enter the Goal Node: ")

all_paths = find_all_paths(graph,start,goal)
print(f"All Possible Paths from {start} to {goal} is ")
for path in all_paths:
    print(" -> ".join(path))
```

```python
#MAX PATH FINDER
directions = [(0,1),(1,1),(-1,1)] # right , right up diagonal , right down diagonal

def isValid(matrix,i,j):
    return 0 <= i <len(matrix) and 0 <= j < len(matrix[0])

def moveNext(index):
    result_dict = {}
    for di,dj in directions:
        ni,nj = di+index[0] , dj+index[1]
        if isValid(matrix,ni,nj):
            result_dict[matrix[ni][nj]] = [ni,nj]
    
    max_value = max(result_dict)
    return result_dict[max_value]
    

def find_max_path(matrix):
    path_cost = 0
    max_value = float('-inf')
    index = []
    for j in range(1):
        for i in range(len(matrix[0])):
           if max_value < matrix[i][j]:
               max_value = matrix[i][j]
               index=[i,j]
    
    path_cost += max_value
    print(matrix[index[0]][index[1]],end="->")
    
    for i in range(len(matrix[0])-1):
        index = moveNext(index)
        path_cost += matrix[index[0]][index[1]]
        print(matrix[index[0]][index[1]],end="->")
        

    return path_cost
    

matrix = [
    [9,10,9],
    [9,9,-1],
    [6,-3,4]
]


maxPath = find_max_path(matrix)
print()
print(f"Maximum Path = {maxPath}")


```

```python
#UCS
def get_min_cost_node(priority_queue):
    """Function to get the node with the lowest cost manually."""
    min_cost_index = 0
    for i in range(len(priority_queue)):
        if priority_queue[i][0] < priority_queue[min_cost_index][0]:
            min_cost_index = i
    return priority_queue.pop(min_cost_index)  # Remove and return the lowest-cost node

def uniform_cost_search(graph, start, goal):
    priority_queue = [(0, start, [])]  # (cost, node, path)
    visited = set()

    while priority_queue:
        cost, node, path = get_min_cost_node(priority_queue)  # Get the node with the least cost

        if node in visited:
            continue
        visited.add(node)

        path = path + [node]

        if node == goal:
            print(f"Path: {' -> '.join(path)}, Cost: {cost}")
            return

        for neighbor, edge_cost in graph.get(node, []):
            if neighbor not in visited:
                priority_queue.append((cost + edge_cost, neighbor, path))  # Add node to queue

    print("No path found")

# Example graph as an adjacency list with edge costs
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('C', 2), ('D', 5)],
    'C': [('D', 1)],
    'D': []
}

start_node = input("Enter the start node: ")
goal_node = input("Enter the goal node: ")

uniform_cost_search(graph, start_node, goal_node)
```

```python
#MAM CODE UCS
#UNIFORM COST SEARCH

import heapq
import matplotlib.pyplot as plt

def uniform_cost_search(graph, start, goal):
    # Priority queue to store the frontier nodes, initialized with the start node
    priority_queue = [(0, start)]
    # Dictionary to store the cost of the shortest path to each node
    visited = {start: (0, None)}
   
    while priority_queue:
        # Pop the node with the lowest cost from the priority queue
        current_cost, current_node = heapq.heappop(priority_queue)
       
        # If we reached the goal, return the total cost and the path
        if current_node == goal:
            return current_cost, reconstruct_path(visited, start, goal)
       
        # Explore the neighbors
        for neighbor, cost in graph[current_node]:
            total_cost = current_cost + cost
            # Check if this path to the neighbor is better than any previously found
            if neighbor not in visited or total_cost < visited[neighbor][0]:
                visited[neighbor] = (total_cost, current_node)
                heapq.heappush(priority_queue, (total_cost, neighbor))
   
    # If the goal is not reachable, return None
    return None
def reconstruct_path(visited, start, goal):
    # Reconstruct the path from start to goal by following the visited nodes
    path = []
    current = goal
    while current is not None:
        path.append(current)
        current = visited[current][1]  # Get the parent node
    path.reverse()
    return path

graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 1), ('E', 3)],
    'C': [('F', 5)],
    'D': [('G', 2)],
    'E': [('G', 1)],
    'F': [('G', 2)],
    'G': []
}

# Example usage of the UCS function
start_node = 'A'
goal_node = 'G'
result = uniform_cost_search(graph, start_node, goal_node)

if result:
    total_cost, path = result
    print(f"Least cost path from {start_node} to {goal_node}: {' -> '.join(path)} with total cost {total_cost}")
   
else:
    print(f"No path found from {start_node} to {goal_node}")
```
